```python
def activation(output, threshold):
    if output >= threshold:
        return 1
    else:
        return 0

def perceptron(orInput):
    a = [0, 0, 1, 1]
    b = [0, 1, 0, 1]
    y = [0, 1, 1, 1]

    if len(a) != len(b) or len(b) != len(y):
        raise ValueError("Panjang input harus sama untuk semua variabel.")

    w = [0, 0]
    threshold = 0.4
    learningRate = 0.5

    inputLength = len(a)
    i = 0

    print("Perceptron Training :")
    print("#####################")
    print("-----------------")

    while i < inputLength:
        error_count = 0

        for i in range(inputLength):
            summation = a[i] * w[0] + b[i] * w[1]
            output = activation(summation, threshold)

            print(f"Input: {a[i]}, {b[i]}")
            print(f"Weights: {w[0]}, {w[1]}")
            print(f"Summation: {summation} | Threshold: {threshold}")
            print(f"Actual Output: {y[i]} | Predicted Output: {output}")
            print("                                                  ")

            if output != y[i]:
                w[0] += learningRate * (y[i] - output) * a[i]
                w[1] += learningRate * (y[i] - output) * b[i]
                error_count += 1

                print("\nWeights Updated, Training Again :")
                print("###################################")

        print("------------------")

        if error_count == 0:
            print("Training selesai")
            break

    summation = orInput[0] * w[0] + orInput[1] * w[1]
    return activation(summation, threshold)

orInput = [1, 1]
print("OR Operation Output For " + str(orInput) + " : " + str(perceptron(orInput)))
```
